```python
import pandas as pd
from datetime import datetime
import os
import sys
```

```python
# reading pdfs using pandas
pd.read_csv('annual-enterprise-survey-2020-financial-year-provisional-csv.csv')
```

```python
# web scraping using selenium

from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.chrome.service import Service

# Preparing script before we convert it to executable
application_path = os.path.dirname(sys.executable)

# get date in format MMDDYYYY
now = datetime.now()
month_day_year = now.strftime("%m%d%Y")


web = 'https://coinmarketcap.com/'
path = '/Users/asus/Downloads/chromedriver'  # introduce path here

# setting up headless mode

options = Options()
options.headless = True

# Creating the driver
driver_service = Service(executable_path=path)
driver = webdriver.Chrome(service=driver_service, options=options)
driver.get(web)

parent_container = driver.find_elements(by='xpath', value='//div[@class="sc-1rmt1nr-0 sc-1rmt1nr-2 iMyvIy"]/*')
trending_container = parent_container[0]
recently_added_container = parent_container[1]

trending_coins_container = trending_container.find_elements(by='xpath', value='./div[@class="sc-1rmt1nr-0 sc-1rmt1nr-4 eQRTPY"]')
recently_added_coins_container = recently_added_container.find_elements(by='xpath', value='./div[@class="sc-1rmt1nr-0 sc-1rmt1nr-4 eQRTPY"]')

trendingCoins = []
recentlyAddedCoins = []
for trending_coin in trending_coins_container:
    rank = trending_coin.find_element(by='xpath', value='.//span[@class="rank"]').text
    coin = trending_coin.find_element(by='xpath', value='.//span[@class="alias"]').text
    logo = trending_coin.find_element(by='xpath', value='.//img').get_attribute("src")
    trendingCoins.append({"rank": rank, "coin": coin, "logo": logo})
    
for recent_coin in recently_added_coins_container:
    rank = recent_coin.find_element(by='xpath', value='.//span[@class="rank"]').text
    coin = recent_coin.find_element(by='xpath', value='.//span[@class="alias"]').text
    logo = recent_coin.find_element(by='xpath', value='.//img').get_attribute("src")
    recentlyAddedCoins.append({"rank": rank, "coin": coin, "logo": logo})
    
my_dict = {'trendingCoins': trendingCoins, 'recentlyAddedCoins': recentlyAddedCoins,}
df_headlines = pd.DataFrame(my_dict)
file_name = f'football_headlines_{month_day_year}.csv'
final_path = os.path.join(application_path, file_name)
df_headlines.to_csv(final_path)
print("hello")
driver.quit()

```
